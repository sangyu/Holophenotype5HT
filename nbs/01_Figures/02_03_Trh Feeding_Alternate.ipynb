```python
# Early and late speed differences
setFont('inter', 10)

# --- Build per-fly mean speed for E and L ---
genoMap = {'trh-gal4>uas-chrimson': 'Trhn>Chr', 'w1118; uas-chrimson': 'Chr/w', 'w1118; trh-gal4': 'Trhn/w'}

resultsDf = TrhCsCh_ele.resultsDf
countLogDf = TrhCsCh_ele.countLogDf

rows = []
for _, fly in resultsDf.iterrows():
    date = fly['Date']
    ch = int(fly['ID'])
    sec_col = date + '_Seconds'
    v_col = f"{date}_Ch{ch}_Obj1_V"

    if sec_col not in countLogDf.columns or v_col not in countLogDf.columns:
        continue

    secs = countLogDf[sec_col]
    speed = countLogDf[v_col]

    m1 = speed[(secs >= 0) & (secs < 3600)].mean()
    m2 = speed[(secs >= 3600) & (secs < 7200)].mean()

    base = {'Status': fly['Status'], 'Genotype': fly['Genotype'],
            'Temperature': fly['Temperature'], 'Date': date, 'ID': v_col}
    rows.append({**base, 'Half': 'E',  'MeanSpeed': m1})
    rows.append({**base, 'Half': 'L', 'MeanSpeed': m2})

speedHalfDf = pd.DataFrame(rows)
speedHalfDf['GenoReduced'] = speedHalfDf['Genotype'].map(genoMap).astype('category').cat.set_categories(['Chr/w', 'Trhn/w', 'Trhn>Chr'], ordered=True)
speedHalfDf['MeanSpeed'] = speedHalfDf['MeanSpeed']  # convert to mm/s if needed
speedHalf_ct_Df = speedHalfDf.loc[speedHalfDf.Status == 'Ctrl']
speedHalf_ts_Df = speedHalfDf.loc[speedHalfDf.Status == 'Test']

f_speed_two_halves, axes_speed_two_halves = plt.subplots(1, 2)
f_speed_two_halves.set_size_inches(10, 6)
f_speed_two_halves.subplots_adjust(wspace=0.6)

cSpeed_ct = dabest.load(speedHalf_ct_Df,
                     x=['Temperature', 'GenoReduced'],
                     y='MeanSpeed',
                     delta2=True,
                     experiment='Half',
                     experiment_label=['E', 'L'],
                     x1_level=['Red Light Off', 'Red Light On'],
                     )

fSpeed_ct = cSpeed_ct.mean_diff.plot(
    contrast_bars_kwargs={'alpha': .15},
    raw_marker_size=0.4, contrast_marker_size=5, raw_desat=1,
    color_col='GenoReduced', custom_palette=trhcschpalette,
    fig_size=(5, 4),
    raw_ylim=(-.05, 5),
    contrast_ylim=(-1, 1.5),
    raw_label='Mean Speed (mm/s)',
    contrast_label='Δ Mean Speed (mm/s)',
    delta2_label='ΔΔ Mean Speed (mm/s)',
    fontsize_contrastxlabel=contrastTextFontSize,
    fontsize_contrastylabel=contrastTextFontSize,
    fontsize_delta2label=contrastTextFontSize,
    fontsize_rawxlabel=contrastTextFontSize,
    fontsize_rawylabel=contrastTextFontSize,
    ax=axes_speed_two_halves[0]
)
cSpeed_ts = dabest.load(speedHalf_ts_Df,
                     x=['Temperature', 'GenoReduced'],
                     y='MeanSpeed',
                     delta2=True,
                     experiment='Half',
                     experiment_label=['E', 'L'],
                     x1_level=['Red Light Off', 'Red Light On'],
                     )

fSpeed_ts = cSpeed_ts.mean_diff.plot(
    contrast_bars_kwargs={'alpha': .15},
    raw_marker_size=0.4, contrast_marker_size=5, raw_desat=1,
    color_col='GenoReduced', custom_palette=trhcschpalette,
    fig_size=(5, 4),
    raw_ylim=(-.05, 5),
    contrast_ylim=(-1, 1.5),
    raw_label='Mean Speed (mm/s)',
    contrast_label='Δ Mean Speed (mm/s)',
    delta2_label='ΔΔ Mean Speed (mm/s)',
    fontsize_contrastxlabel=contrastTextFontSize,
    fontsize_contrastylabel=contrastTextFontSize,
    fontsize_delta2label=contrastTextFontSize,
    fontsize_rawxlabel=contrastTextFontSize,
    fontsize_rawylabel=contrastTextFontSize,
    ax = axes_speed_two_halves[1],
)

order = [1,0,2]
axes_speed_two_halves[0].legend(['Chr/w', 'Trhn/w'], frameon=True,loc='right', bbox_to_anchor=[1.1, 1.1],  fontsize= contrastTextFontSize, markerscale = 1)
leg_ct = axes_speed_two_halves[0].get_legend()
hl_dict = {handle.get_label(): handle for handle in leg_ct.legend_handles}
hl_dict['_child0'].set_color(trhcschpalette['Trhn/w'])
hl_dict['_child1'].set_color(trhcschpalette['Chr/w'])

axes_speed_two_halves[1].legend(['Trhn>Chr'], frameon=True,loc='right', bbox_to_anchor=[1.1, 1.1],  fontsize= contrastTextFontSize, markerscale = 1)
leg_ts = axes_speed_two_halves[1].get_legend()
hl_dict = {handle.get_label(): handle for handle in leg_ts.legend_handles}
hl_dict['_child0'].set_color(trhcschpalette['Trhn>Chr'])

axes_speed_two_halves[0].set_xticklabels(['Off', 'On', 'Off', 'On'])
axes_speed_two_halves[0].contrast_axes.set_xticklabels(['', 'Early', '', 'Late',  'ΔΔ'])
axes_speed_two_halves[1].set_xticklabels(['Off', 'On', 'Off', 'On'])
axes_speed_two_halves[1].contrast_axes.set_xticklabels(['', 'Early', '', 'Late',  'ΔΔ'])
# f_speed_two_halves.suptitle('Trhn>Chr Speed: Early vs Late',
#                 font='inter', style='italic', weight='medium', fontsize=14, color='steelblue', y=.97)
f_speed_two_halves.savefig(figurepath + 'TrhCsCh_SpeedHalf_contrast.svg')



make_delta2_results(cSpeed_ct, 'Early', 'Late').to_csv(figurepath + 'TrhCsCh_SpeedHalf_ctrl_contrast.csv')
make_delta2_results(cSpeed_ts, 'Early', 'Late').to_csv(figurepath + 'TrhCsCh_SpeedHalf_test_contrast.csv')
```
