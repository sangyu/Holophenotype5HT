```python
# Patch svgutils.SVG to handle SVGs that have viewBox but no width/height attributes
import svgutils.transform as _st
from svgutils.compose import SVG as _SVG, Element as _Element

def _patched_svg_init(self, fname, fix_mpl=False):
    svg = _st.fromfile(fname)
    if svg.width is None:
        vb = svg.root.get('viewBox')
        if vb:
            parts = vb.replace(',', ' ').split()
            svg.root.set('width', parts[2])
            svg.root.set('height', parts[3])
    self._width = None if (svg.width is None or svg.width.endswith("%")) else svg.width
    self._height = None if (svg.height is None or svg.height.endswith("%")) else svg.height
    _Element.__init__(self, svg.getroot().root)

_SVG.__init__ = _patched_svg_init
```
